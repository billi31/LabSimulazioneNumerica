# Laboratorio 6
## Esercizio 06.1
Di seguito mostro i risultati per un modello di Ising monodimensionale di:
1. $U(N,T) = \langle H \rangle_{h=0}$;
2. $C(N,T) = k_B \beta^2 (\langle H^2 \rangle_{h=0}-\langle H \rangle_{h=0}^2)$;
3. $\chi(N,T) = \beta \langle (\sum_{i=1,N}s_i)^2 \rangle_{h=0}$;
4. $M(N,T,h=0.02)= \langle \sum_{i=1,N}s_i \rangle_{h=0.02}$;

ottenute campionando la distribuzione di probabilit√† di Boltzmann tramite algoritmo di Metropolis e di Gibbs. Per la statistica ho simulato $n_{step}=10000$ step che ho suddiviso in $n_{block}=100$ blocchi per ogni temperatura considerata ($T\in[0.5,2]$). Ho simulato un modello di Ising con $n_{spin}=50$. Ho svolto una prima parte di equilibrazione effettuando $n_{equil}=100$ step.

```python
temp_m, ene_temperatura_m, err_ene_temperatura_m = np.loadtxt("metro/output0/output.ene.temp.0", unpack=True)
temp_g, ene_temperatura_g, err_ene_temperatura_g = np.loadtxt("gibbs/output0/output.ene.temp.0", unpack=True)
temp_m, heat_temperatura_m, err_heat_temperatura_m = np.loadtxt("metro/output0/output.heat.temp.0", unpack=True)
temp_g, heat_temperatura_g, err_heat_temperatura_g = np.loadtxt("gibbs/output0/output.heat.temp.0", unpack=True)
temp_m, chi_temperatura_m, err_chi_temperatura_m = np.loadtxt("metro/output0/output.chi.temp.0", unpack=True)
temp_g, chi_temperatura_g, err_chi_temperatura_g = np.loadtxt("gibbs/output0/output.chi.temp.0", unpack=True)
temp_m, mag_temperatura_m, err_mag_temperatura_m = np.loadtxt("metro/output0.2/output.mag.temp.0", unpack=True)
temp_g, mag_temperatura_g, err_mag_temperatura_g = np.loadtxt("gibbs/output0.2/output.mag.temp.0", unpack=True)

fig = plt.figure(figsize = (18,18))

points=100
T = np.linspace(0.5,2.0,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )

fig.suptitle("Thermodynamic Quantities, Ising 1D", fontsize = 16, fontweight = "bold")
ax = fig.add_subplot(221)
ax.plot(T, e, label = "Internal Energy")
ax.set_title('Internal Energy', fontsize = 14, fontweight = "bold")
ax.set_xlabel('T', fontsize = 12, fontweight = "bold")
ax.set_ylabel('U/N', fontsize = 12, fontweight = "bold")

ax.errorbar(temp_m,ene_temperatura_m,yerr=err_ene_temperatura_m, label = "Metropolis", marker = '.')
ax.errorbar(temp_g,ene_temperatura_g,yerr=err_ene_temperatura_g, label = "Gibbs", marker = '.')
ax.legend()
ax.grid()

ax = fig.add_subplot(222)
heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)

ax.plot(T, heat, label = "Heat Capacity")
ax.set_title('Heat Capacity', fontsize = 14, fontweight = "bold")
ax.set_xlabel('T', fontsize = 12, fontweight = "bold")
ax.set_ylabel('C', fontsize = 12, fontweight = "bold")

ax.errorbar(temp_m,heat_temperatura_m,yerr=err_heat_temperatura_m, label = "Metropolis", marker = '.')
ax.errorbar(temp_g,heat_temperatura_g,yerr=err_heat_temperatura_g, label = "Gibbs", marker = '.')

ax.legend()
ax.grid()

ax = fig.add_subplot(223)
X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)


ax.plot(T, X, label = "Magnetic Susceptibility")
ax.set_title('Magnetic Susceptibility', fontsize = 14, fontweight = "bold")
ax.set_xlabel('T', fontsize = 12, fontweight = "bold")
ax.set_ylabel('X', fontsize = 12, fontweight = "bold")


ax.errorbar(temp_m,chi_temperatura_m,yerr=err_chi_temperatura_m, label = "Metropolis", marker = '.')
ax.errorbar(temp_g,chi_temperatura_g,yerr=err_chi_temperatura_g, label = "Gibbs", marker = '.')

ax.legend()
ax.grid()

ax = fig.add_subplot(224)

b=1/T
h=0.2
l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)
ax.plot(T, M, label = "Magnetization")


ax.set_title('Magnetization', fontsize = 14, fontweight = "bold")
ax.set_xlabel('T', fontsize = 12, fontweight = "bold")
ax.set_ylabel('M', fontsize = 12, fontweight = "bold")


ax.errorbar(temp_m,mag_temperatura_m,yerr=err_mag_temperatura_m, label = "Metropolis", marker = '.')
ax.errorbar(temp_g,mag_temperatura_g,yerr=err_mag_temperatura_g, label = "Gibbs", marker = '.')

ax.legend()
ax.grid()
plt.show()
```

Nei grafici sono riportati i risultati finali ottenuti dalla media a blocchi: le simulazioni conordano pienamente con gli andamenti teorici. Mostro ora l'andamento degli errori trovati in funzione della temperature:

```python
fig = plt.figure(figsize = (17,12))

fig.suptitle("Thermodynamic Quantities Error, Ising 1D", fontsize = 16, fontweight = "bold")


ax = fig.add_subplot(221)
ax.set_title('Internal Energy Error', fontsize = 14, fontweight = "bold")
ax.set_xlabel('T', fontsize = 12, fontweight = "bold")
ax.set_ylabel('U/N', fontsize = 12, fontweight = "bold")

ax.plot(temp_m,err_ene_temperatura_m, label = "Metropolis", marker = '.', color = "tab:orange")
ax.plot(temp_g,err_ene_temperatura_g, label = "Gibbs", marker = '.', color = "tab:green")
ax.legend()
ax.grid()

ax = fig.add_subplot(222)
ax.set_title('Heat Capacity Error', fontsize = 14, fontweight = "bold")
ax.set_xlabel('T', fontsize = 12, fontweight = "bold")
ax.set_ylabel('C', fontsize = 12, fontweight = "bold")

ax.plot(temp_m,err_heat_temperatura_m, label = "Metropolis", marker = '.', color = "tab:orange")
ax.plot(temp_g,err_heat_temperatura_g, label = "Gibbs", marker = '.', color = "tab:green")
ax.legend()
ax.grid()

ax = fig.add_subplot(223)
ax.set_title(r'Magnetic Susceptibility Error',fontsize = 14, fontweight = "bold")
ax.set_xlabel(r'T', fontweight = "bold", fontsize = 12)
ax.set_ylabel(r'X', fontweight = "bold", fontsize = 12)

ax.plot(temp_m,err_chi_temperatura_m, label = "Metropolis", marker = '.', color = "tab:orange")
ax.plot(temp_g,err_chi_temperatura_g, label = "Gibbs", marker = '.', color = "tab:green")
ax.legend()
ax.grid()

ax = fig.add_subplot(224)
ax.set_title('Magnetization Error', fontsize = 14, fontweight = "bold")
ax.set_xlabel('T', fontsize = 12, fontweight = "bold")
ax.set_ylabel('M', fontsize = 12, fontweight = "bold")

ax.plot(temp_m,err_mag_temperatura_m, label = "Metropolis", marker = '.', color = "tab:orange")
ax.plot(temp_g,err_mag_temperatura_g, label = "Gibbs", marker = '.', color = "tab:green")
ax.legend()
ax.grid()


plt.show()
```

In generale sembra che l'algoritmo di Metropolis abbia un errore minore dell'algoritmo di Gibbs.
