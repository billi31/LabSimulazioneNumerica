## Esercizio 09.1
Per risolvere il problema del commesso viaggiatore ho effettuato una rapida grid search per determinare quanti individui e quante generazione fossero ottimali. Di seguito mostro i risultati ottenuti per:
- 32 città disposte su una circonferenza e, separatamente, 32 città disposte in un quadrato;
- funzione di costo $L^{(1)}$.

In ogni nuova generazione ho scelto gli individui tramite una funzione che favorisse gli individui della generazione precedente con una funzione di costo minore; ho effettuato un crossover con probabilità del $60\%$ e una mutazione (ce ne sono 6 possibili) con probabilità del $10\%$.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import scipy.optimize as opt
import cv2
```

```python
i_individuals_circle, i_generations_circle, L1_grid_circle = np.loadtxt('output/circle/L1/grid.dat', unpack=True)
i_individuals_square, i_generations_square, L1_grid_square = np.loadtxt('output/square/L1/grid.dat', unpack=True)

i_individuals_circle = np.reshape(i_individuals_circle, (20, -1))
i_individuals_circle = i_individuals_circle.T
i_generations_circle = np.reshape(i_generations_circle, (-1, 10))
L1_grid_circle = np.reshape(L1_grid_circle, (-1, 10))
i_individuals_square = np.reshape(i_individuals_square, (20, -1))
i_individuals_square = i_individuals_square.T
i_generations_square = np.reshape(i_generations_square, (-1, 10))
L1_grid_square = np.reshape(L1_grid_square, (-1, 10))

fig = plt.figure(figsize = (18,8))
fig.suptitle('Final $L^{(1)}$ evaluated for different values of $\mathbf{n_{generations}}$ and $\mathbf{n_{individuals}}$', fontsize = 16, fontweight = "bold")

ax = fig.add_subplot(121)
#x = generations, y = individuals
image = ax.imshow(L1_grid_circle, extent = [200 , 700 , 250 , 50], aspect = 'auto')
plt.colorbar(image, label = "Final L1")
ax.set_xlabel('$\mathbf{n_{generations}}$', fontsize = 12, fontweight = 'bold')
ax.set_ylabel('$\mathbf{n_{individuals}}$', fontsize = 12, fontweight = 'bold')
ax.set_title('Cities on a Circle', fontsize = 14, fontweight = 'bold')

ax = fig.add_subplot(122)
#x = generations, y = individuals
image = ax.imshow(L1_grid_square, extent = [400 , 900 , 300 , 100], aspect='auto')
plt.colorbar(image, label = "Final L1")
ax.set_xlabel('$\mathbf{n_{generations}}$', fontsize = 12, fontweight = 'bold')
ax.set_ylabel('$\mathbf{n_{individuals}}$', fontsize = 12, fontweight = 'bold')
ax.set_title('Cities in a Square', fontsize = 14, fontweight = 'bold')



plt.show()

ij_min_circle = np.where(L1_grid_circle == np.amin(L1_grid_circle))
print('CIRCLE:')
print("Indexes of minimum L1",ij_min_circle)
print("n_individuals of minimum L1",i_individuals_circle[0][ij_min_circle[0][0]])
print("n_generations of minimum L1",i_generations_circle[0][ij_min_circle[1][0]])
print("Minimum L1",np.amin(L1_grid_circle))
ij_min_square = np.where(L1_grid_square == np.amin(L1_grid_square))
print('SQUARE:')
print("Indexes of minimum L1",ij_min_square)
print("n_individuals of minimum L1",i_individuals_square[0][ij_min_square[0][0]])
print("n_generations of minimum L1",i_generations_square[0][ij_min_square[1][0]])
print("Minimum L1",np.amin(L1_grid_square))
```

Mi concentro sui minimi appena trovati. Per la simulazione del cerchio continuerò con $n_{individuals}=230$ e $n_{generations}=300$, mentre per la simulazione del quadrato continuerò con $n_{individuals}=180$ e $n_{generations}=500$. Mostro la soluzione trovata:

```python
icity_circle, x_circle, y_circle = np.loadtxt('output/circle/L1/solution.dat', unpack=True)
icity_square, x_square, y_square = np.loadtxt('output/square/L1/solution.dat', unpack=True)

fig = plt.figure(figsize = (18,8))
fig.suptitle('Best Path Found', fontsize = 16, fontweight = "bold")

ax = fig.add_subplot(121)
ax.set_xlabel('$\mathbf{x}$', fontsize = 12, fontweight = 'bold')
ax.set_ylabel('$\mathbf{y}$', fontsize = 12, fontweight = 'bold')
ax.set_title('$\mathbf{n_{generations}}=300$,$\mathbf{n_{individuals}}=230$', fontsize = 14, fontweight = 'bold')
ax.plot(x_circle, y_circle, marker = "x")
ax = fig.add_subplot(122)
ax.set_xlabel('$\mathbf{x}$', fontsize = 12, fontweight = 'bold')
ax.set_ylabel('$\mathbf{y}$', fontsize = 12, fontweight = 'bold')
ax.set_title('$\mathbf{n_{generations}}=500$,$\mathbf{n_{individuals}}=180$', fontsize = 14, fontweight = 'bold')
ax.plot(x_square, y_square, marker = "x")

plt.show()
```

Mostro ora i grafici delle funzioni di costo: confronto il valore della funzione di costo del miglior percorso di ogni generazione con la media della funzione di costo per la metà migliore dei percorsi di ogni generazione.

```python
L1_circle = np.loadtxt('output/circle/L1/best.fitness.dat', unpack=True)
L1_ave_circle = np.loadtxt('output/circle/L1/ave.fitness.dat', unpack=True)
L1_square = np.loadtxt('output/square/L1/best.fitness.dat', unpack=True)
L1_ave_square = np.loadtxt('output/square/L1/ave.fitness.dat', unpack=True)

fig = plt.figure(figsize = (18,8))
fig.suptitle('Cost Function $\mathbf{L^{(1)}}$', fontsize = 16, fontweight = "bold")

ax = fig.add_subplot(121)
ax.set_xlabel('$\mathbf{n_{generations}}$', fontsize = 12, fontweight = 'bold')
ax.set_ylabel('$\mathbf{L^{(1)}}$', fontsize = 12, fontweight = 'bold')
ax.set_title('Circle', fontsize = 14, fontweight = 'bold')
ax.plot(L1_circle, label = 'L1 of Best Individual')
ax.plot(L1_ave_circle, label = 'L1 Average in Half Population')
ax.legend()
ax.grid()

ax = fig.add_subplot(122)
ax.set_xlabel('$\mathbf{n_{generations}}$', fontsize = 12, fontweight = 'bold')
ax.set_ylabel('$\mathbf{L^{(1)}}$', fontsize = 12, fontweight = 'bold')
ax.set_title('Square', fontsize = 14, fontweight = 'bold')
ax.plot(L1_square, label = 'L1 of Best Individual')
ax.plot(L1_ave_square, label = 'L1 Average in Half Population')
ax.legend()
ax.grid()

plt.show()

print('CIRCLE Final L1:    ',L1_circle[len(L1_circle)-1])
print('SQUARE Final L1:   ',L1_square[len(L1_square)-1])
```

Dai grafici della funzione di costo vediamo come i diversi percorsi considerati migliorino progressivamente. All'avvicinarsi alla soluzione si vede come i percorsi tendano ad uniformarsi al percorso migliore, fino ad arrivare ad una convergenza.
