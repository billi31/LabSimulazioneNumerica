## Esercizio 02.1
Valuto il valore dell'integrale 
$$
I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1
$$
utilizzando un generatore di numeri random distribuiti uniformemente e confronto i risultati ottenuti tramite importance sampling. La funzione che ho utilizzato per l'importance sampling Ã¨ la retta che intercetta gli assi in $y=\pi/2$ e $x=1$.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.optimize as opt
sns.set()
```

```python
x, sum_prog, err_prog = np.loadtxt('output/int_unif.txt', unpack=True)
x_IS, sum_prog_IS, err_prog_IS = np.loadtxt('output/int_cos_IS.txt', unpack=True)
sum_prog = sum_prog - 1
sum_prog_IS = sum_prog_IS - 1
```

```python
fig = plt.figure(figsize = (17,5))
ax = fig.add_subplot(121)
ax.errorbar(x,sum_prog,yerr=err_prog)
ax.set_title(r'Integral Extimation with Uniform Distribution',fontsize = 14, fontweight = "bold")

ax.set_xlabel(r'# of Block', fontweight = "bold")
ax.set_ylabel(r'$\mathbf{I - 1}$')
ax.grid(True)

ax = fig.add_subplot(122)
ax.errorbar(x_IS,sum_prog_IS,yerr=err_prog_IS)
ax.set_title(r'Integral Extimation with Importance Sampling',fontsize = 14, fontweight = "bold")

ax.set_xlabel(r'# of Block', fontweight = "bold")
ax.set_ylabel(r'$\mathbf{I - 1}$')
ax.grid(True)

plt.show()
```

Dai due grafici vediamo la convergenza dell'integrale con entrambi i metodi. L'errore valutato con l'importance sampling Ã¨ minore di quello ottenuto con il primo metodo:

```python
fig, axes = plt.subplots(figsize = (9,5))
plt.plot(x,err_prog, label = "Uniform Distribution")
plt.plot(x_IS,err_prog_IS,'r', label = "Importance Sampling")

plt.title(r'Integral Extimation Error',fontsize = 14, fontweight = "bold")

plt.xlabel('# of Block', fontweight = "bold")
plt.ylabel('Error', fontweight = "bold")
plt.legend()
plt.show()


print("Mean Final Error: {:0.5f}".format(err_prog[(len(x)-1)]))
print("Variance Final Error: {:0.6f}".format(err_prog_IS[(len(x_IS)-1)]))
```

Come ci si puÃ² aspettare, utilizzando l'importance sampling (con una funzione che si avvicina alla funzione integranda) l'errore calcolato Ã¨ molto minore.

## Esercizio 02.2
Procedo simulando $M=100000$ random walk (sia nel reticolo cubico, sia nel continuo) per ottenere $\sqrt{\langle r^2_i\rangle}$, dove $i\in[0,100]$ Ã¨ il numero di passi eseguiti. Per valutarne la statistica suddivido in $N=100$ blocchi. 

```python
x_d, sum_prog_d, err_d = np.loadtxt('output/RW_d.txt', unpack=True)
x_c, sum_prog_c, err_c = np.loadtxt('output/RW_c.txt', unpack=True)
sum_prog_d = np.sqrt(sum_prog_d)
err_d = [0 if (sum_prog_d[i] == 0) else err_d[i]/(2*sum_prog_d[i]) for i in range(len(err_d))]

sum_prog_c = np.sqrt(sum_prog_c)
err_c = [0 if (sum_prog_c[i] == 0) else err_c[i]/(2*sum_prog_c[i]) for i in range(len(err_c))]
```

```python
def fit(N,k):
    return k*np.sqrt(N)
opt_param_d, pcov_d = opt.curve_fit(fit,x_d,sum_prog_d)
opt_param_c, pcov_c = opt.curve_fit(fit,x_c,sum_prog_c)
print("Fit Parameter")
print("Cubic Lattice RW : {0:0.5f}".format(opt_param_d[0]))

print("Continuous RW    : {0:0.5f}".format(opt_param_c[0]))
```

```python
fig, axes = plt.subplots(figsize = (14,8))
plt.errorbar(x_d,sum_prog_d,yerr=err_d, linestyle="", marker=".", label="Simulation")

plt.title(r'Random Walk in Cubic Lattice',fontsize = 14, fontweight = "bold")

plt.xlabel('# of Steps', fontweight = "bold")
plt.ylabel('$\mathbf{\sqrt{<r_n^2>}}$')
plt.grid(True)

plt.plot(x_d,fit(x_d,opt_param_d), 'r', label = "Fit")
plt.legend()
plt.show()

fig, axes = plt.subplots(figsize = (14,8))
plt.errorbar(x_c,sum_prog_c,yerr=err_c, linestyle="", marker=".", label="Simulation")

plt.title(r'Random Walk in Continuum',fontsize = 14, fontweight = "bold")

plt.xlabel('# of Steps', fontweight = "bold")
plt.ylabel('$\mathbf{\sqrt{<r_n^2>}}$')
plt.grid(True)

plt.plot(x_c,fit(x_c,opt_param_c), 'r', label = "Fit")
plt.legend()
plt.show()
```

Dai grafici Ã¨ evidente un comportamento diffusivo in entrambi i casi: la distanza media raggiunta Ã¨ proporzionale a $\sqrt{n_{step}}$.
