## Esercizio 10.1
Ricerco la soluzione del problema del commesso viaggiatore tramite Simulated Annealing. Questo tipo di algoritmo si ispira al processo fisico di cristallizzazione per sistemi raffreddati molto lentamente, che seguono la distribuzione di Boltzmann
$$p(x)=\frac{e^{-\beta E(x)}}{Z}$$
con la quale riescono ad arrivare allo stato fondamentale di energia minima. Negli algoritmi di Simulated Annealing l'energia del sistema viene interpretata come una funzione di costo $L(x)$: il sistema tenderà, al diminuire della temperatura, a "raffreddarsi" nella configurazione con funzione di costo minima. Ad ogni step di temperatura si effettuano un numero di step Montecarlo che permetta al percorso di ridistribuirsi.

Il codice diminuisce progressivamente la temperatura, partendo da $T=1.5$ e riducendola dello $0,05\%$ in ogni step ($n_{temp\ step}=2000$ step); per ogni temperatura sono stati svolti $n_{MC}=700$ step dell'algoritmo Metropolis. La funzione di costo utilizzata è $L^{(1)}$.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
icity_circle, x_circle, y_circle = np.loadtxt('output/circle/L1/solution.dat', unpack=True)
icity_GA_circle, x_GA_circle, y_GA_circle = np.loadtxt('../9lab/output/circle/L1/solution.dat', unpack=True)
icity_square, x_square, y_square = np.loadtxt('output/square/L1/solution.dat', unpack=True)
icity_GA_square, x_GA_square, y_GA_square = np.loadtxt('../9lab/output/square/L1/solution.dat', unpack=True)


fig = plt.figure(figsize = (18,18))
fig.suptitle('Best Path Found', fontsize = 16, fontweight = "bold")

ax = fig.add_subplot(221)
ax.set_xlabel('$\mathbf{x}$', fontsize = 12, fontweight = 'bold')
ax.set_ylabel('$\mathbf{y}$', fontsize = 12, fontweight = 'bold')
ax.set_title('Simulated Annealing', fontsize = 14, fontweight = 'bold')
ax.plot(x_circle, y_circle, marker = "x")
ax = fig.add_subplot(222)
ax.set_xlabel('$\mathbf{x}$', fontsize = 12, fontweight = 'bold')
ax.set_ylabel('$\mathbf{y}$', fontsize = 12, fontweight = 'bold')
ax.set_title('Genetic Algorithm', fontsize = 14, fontweight = 'bold')
ax.plot(x_GA_circle, y_GA_circle, marker = "x")

ax = fig.add_subplot(223)
ax.set_xlabel('$\mathbf{x}$', fontsize = 12, fontweight = 'bold')
ax.set_ylabel('$\mathbf{y}$', fontsize = 12, fontweight = 'bold')
ax.set_title('Simulated Annealing', fontsize = 14, fontweight = 'bold')
ax.plot(x_square, y_square, marker = "x")

ax = fig.add_subplot(224)
ax.set_xlabel('$\mathbf{x}$', fontsize = 12, fontweight = 'bold')
ax.set_ylabel('$\mathbf{y}$', fontsize = 12, fontweight = 'bold')
ax.set_title('Genetic Algorithm', fontsize = 14, fontweight = 'bold')
ax.plot(x_GA_square, y_GA_square, marker = "x")
plt.show()
```

I risultati ottenuti sono buoni per le città disposte sulla circonferenza; per le città disposte nel quadrato, invece, si può già immaginare che la soluzione sia leggermente peggiore di quella ottenuta tramite l'algoritmo genetico.

Anche nel caso delle città disposte sulla circonferenza si vede come l'algoritmo genetico riesca ad ottenere una soluzione più vantaggiosa. Ho notato come il codice da me sviluppato non riesca a trovare facilmente la soluzione del problema del commesso viaggiatore, anche nel caso semplice di città su una circonferenza: è estremamente sensibile al numero di step Montecarlo simulati e tende a "congelarsi" spesso in configurazioni con bassa funzione di costo, ma difficilmente riesce ad arrivare al minimo globale.

## Esercizio 10.2
Riprendo il codice dell'esercizio 09.1 e lo adatto alla programmazione in parallelo, simulando quattro ricerche parallele. Ogni $N_{migr}=25$ generazioni di individui i processi si scambiano a due a due i due percorsi migliori che hanno sviluppato.

```python
icity_square, x_square, y_square = np.loadtxt('MPI/output/square/L1/0/solution.dat', unpack=True)

plt.figure(figsize = (8,8))

plt.xlabel('$\mathbf{x}$', fontsize = 14, fontweight = 'bold')
plt.ylabel('$\mathbf{y}$', fontsize = 14, fontweight = 'bold')
plt.title('Best Path Found', fontsize = 16, fontweight = 'bold')
plt.plot(x_square, y_square, marker = "x")


plt.show()
```

Si nota come la soluzione ottenuta non sia ottimale, vista la presenza di un incrocio nel percorso. Mostro l'andamento della funzione di costo:

```python
L1_square_0 = np.loadtxt('MPI/output/square/L1/0/best.fitness.dat', unpack=True)
L1_square_1 = np.loadtxt('MPI/output/square/L1/1/best.fitness.dat', unpack=True)
L1_square_2 = np.loadtxt('MPI/output/square/L1/2/best.fitness.dat', unpack=True)
L1_square_3 = np.loadtxt('MPI/output/square/L1/3/best.fitness.dat', unpack=True)
L1_square_rank = np.loadtxt('MPI/output/square/L1/best.fitness.rank.dat', unpack=True)

plt.figure(figsize = (18,8))

plt.xlabel('$\mathbf{n_{generations}}$', fontsize = 14, fontweight = 'bold')
plt.ylabel('$\mathbf{L^{(1)}}$', fontsize = 14, fontweight = 'bold')
plt.title('$\mathbf{L^{(1)}}$ Value per Generation', fontsize = 16, fontweight = 'bold')
plt.plot(L1_square_0, label = 'L1 of Best Individual of rank 0')
plt.plot(L1_square_1, label = 'L1 of Best Individual of rank 1')
plt.plot(L1_square_2, label = 'L1 of Best Individual of rank 2')
plt.plot(L1_square_3, label = 'L1 of Best Individual of rank 3')
plt.plot(L1_square_rank, label = 'L1 of Best Individual', linewidth=2)
plt.grid()
plt.legend()
plt.show()

print('Final Best L1:   ',L1_square_rank[len(L1_square_rank)-1])
```

Dalla ricerca in parallelo ho estratto il valore della funzione di costo del miglior individuo di ogni processo per ogni generazione. Si vede come l'utilizzo di una ricerca in parallelo migliori le prestazioni, ma non in maniera estremamente rilevante.
