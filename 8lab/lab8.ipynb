In tutte e quattro le situazioni evidenziate otteniamo un valore che si discosta di meno di una deviazione standard dall'energia di Ground State che ci aspettiamo. Effettuo, quindi, una generazione di $n=5\times 10^5$ punti distribuiti secondo la probabilità di prova $|\Psi_T^{\sigma,\mu}(x)|^2$, con $\mu=0.795$ e $\sigma=0.590$:

```python
psi2_histo, etot_insta = np.loadtxt('output/instantaneus.psi2_etot.0', unpack=True)

plt.figure(figsize = (12,7))
plt.grid(True)

points=100
x = np.linspace(-3.5,3.5,num=points)
mu = 0.795
sigma = 0.590

e = (np.exp(-(x+mu)**2/(2*sigma**2))+np.exp(-(x-mu)**2/(2*sigma**2)))**2
norm = 2*sigma*np.sqrt(np.pi)*(1.+np.exp(-mu**2/(sigma**2))) 
psi2 = e/norm

plt.hist(psi2_histo,bins = 500, density = True, range = [-3.5,3.5], label = "Monte Carlo", alpha = 0.8, color = 'mediumaquamarine')
plt.plot(x, psi2, label = "Analitic", color = 'crimson')

def Vpot(x):
    return (x**2 - 2.5)*x**2

hbar = 1
m = 1
a = 7
N = 1000 # number of iterations
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)
CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)
E,psi = np.linalg.eigh(H)
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)
plt.plot(x,(psi[0])**2, label = "Numerical", color = 'navy')

plt.xlabel('x', fontsize = 12, fontweight = 'bold')
plt.ylabel('$\mathbf{|\Psi_T(x)|^2}$', fontsize = 12, fontweight = 'bold')
plt.title('Probability Distribution Function $\mathbf{|\Psi(x)|^2}$', fontsize = 16, fontweight = 'bold')
plt.legend()
plt.show()
```

La distribuzione di punti segue la funzione analitica $|\Psi_T^{\sigma,\mu}(x)|^2$, ma, nonostante i tentativi di ottimizzazione, non coincide in maniera ottimale con la soluzione numerica dell'equazione di Schroedinger.

Li confronto anche con i risultati forniti dall'algoritmo PIGS con parte immaginaria $ITP=0.1$:

```python
x_PIGS_constant, psi_PIGS_constant, error_PIGS_constant = np.loadtxt("PIMC_PIGS/pigs/constant/probability.dat", unpack = True)
x_PIGS_variational, psi_PIGS_variational, error_PIGS_variational = np.loadtxt("PIMC_PIGS/pigs/variational/probability.dat", unpack = True)
x_PIGS_reduced, psi_PIGS_reduced, error_PIGS_reduced = np.loadtxt("PIMC_PIGS/pigs/rip/probability.dat", unpack = True)


plt.figure(figsize = (15,10))
plt.grid(True)

points=100
x = np.linspace(-3.5,3.5,num=points)
mu = 0.795
sigma = 0.590

e = (np.exp(-(x+mu)**2/(2*sigma**2))+np.exp(-(x-mu)**2/(2*sigma**2)))**2
norm = 2*sigma*np.sqrt(np.pi)*(1.+np.exp(-mu**2/(sigma**2))) 
psi2 = e/norm

plt.plot(x, psi2, label = "Analitic", color = 'crimson'
        )

def Vpot(x):
    return (x**2 - 2.5)*x**2

hbar = 1
m = 1
a = 7
N = 1000 # number of iterations
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)
CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)
E,psi = np.linalg.eigh(H)
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)
plt.plot(x,(psi[0])**2, label = "Numerical", color = 'navy'
        )
plt.errorbar(x_PIGS_constant,psi_PIGS_constant, yerr=error_PIGS_constant, label = "PIGS constant", color = 'mediumaquamarine'
            )
plt.errorbar(x_PIGS_variational,psi_PIGS_variational, yerr=error_PIGS_variational, label = "PIGS variational", color = 'gold'
            )
plt.errorbar(x_PIGS_reduced,psi_PIGS_reduced, yerr=error_PIGS_reduced, label = "PIGS ITP=0.1", color = 'darkviolet'
            )



plt.xlabel('x', fontsize = 12, fontweight = 'bold')
plt.ylabel('$\mathbf{|\Psi_T(x)|^2}$', fontsize = 12, fontweight = 'bold')
plt.title('Probability Distribution Function $\mathbf{|\Psi(x)|^2}$', fontsize = 16, fontweight = 'bold')
plt.legend()
plt.show()
```

L'algoritmo PIGS sia partendo da una funzione costante, sia partendo dalla funzione varazionale, segue molto bene l'andamento numerico esatto. Nel caso in cui si riduca la parte immaginaria fino a $TIP=0.1$, il risultato si discosta sensibilmente dalla soluzione esatta: mi aspetto che per valori della parte immaginaria ancora minori il risultato si discosti sempre più dalla soluzione.

Per vedere come varia la soluzione all'aumentare della temperatura utilizzo l'algoritmo PIMC (simulato con $T=1.25$):

```python
x_PIMC_variational, psi_PIMC_variational, error_PIMC_variational = np.loadtxt("PIMC_PIGS/pimc/variational/probability.dat", unpack = True)

plt.figure(figsize = (15,10))
plt.grid(True)

points=100
x = np.linspace(-3.5,3.5,num=points)

def Vpot(x):
    return (x**2 - 2.5)*x**2

hbar = 1
m = 1
a = 7
N = 1000 # number of iterations
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)
CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)
E,psi = np.linalg.eigh(H)
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)
plt.plot(x,(psi[0])**2, label = "Numerical", color = 'navy')
plt.errorbar(x_PIMC_variational,psi_PIMC_variational, yerr=error_PIMC_variational, label = "PIMC", color = 'gold')


plt.xlabel('x', fontsize = 12, fontweight = 'bold')
plt.ylabel('$\mathbf{|\Psi_T(x)|^2}$', fontsize = 12, fontweight = 'bold')
plt.title('Probability Distribution Function $\mathbf{|\Psi(x)|^2}$', fontsize = 16, fontweight = 'bold')
plt.legend()
plt.show()
```
