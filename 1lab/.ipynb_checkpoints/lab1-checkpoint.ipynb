## Esercizio 01.1
Per testare il generatore di numeri casuali fornito controlliamo inizialmente i valori di media e deviazione standard. Mi aspetto che un buon generatore di numeri random abbia media
$$
\langle r \rangle = \int_0^1 r dr = \frac12
$$
e deviazione standard
$$
\sigma^2 = \int_0^1 (r-1/2)^2 dr = \frac1{12}.
$$
Per analizzare queste due quantità è necessario valutare l'errore; mi servo, quindi, della media a blocchi, che riesce a fornirne una buona stima. Nel codice ho simulato $n=10^6$ numeri random tra 0 e 1; per la media a blocchi ho utilizzato $n_{blocks}=100$.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import norm, cauchy, expon
import seaborn as sns
sns.set()
```

Importo i risultati ottenuti:

```python
x, sum_prog, err_prog = np.loadtxt('output/mean.txt', unpack=True)
var, var_sum_prog, var_err_prog = np.loadtxt('output/var.txt', unpack=True)
```

```python
sum_prog = sum_prog - 0.5
var_sum_prog = var_sum_prog - 1./12.
```

```python
fig = plt.figure(figsize = (17,5))
ax = fig.add_subplot(121)
ax.errorbar(x,sum_prog,yerr=err_prog)
ax.set_title(r'Mean Extimation of Random Generation',fontsize = 14, fontweight = "bold")

ax.set_xlabel(r'# of Block')
ax.set_ylabel(r'$\langle r\rangle -1/2$')
ax.grid(True)

ax = fig.add_subplot(122)
ax.errorbar(var,var_sum_prog,yerr=var_err_prog)
ax.set_title(r'Variance Extimation of Random Generation',fontsize = 14, fontweight = "bold")

ax.set_xlabel(r'# of Block')
ax.set_ylabel(r'$\sigma^2-1/12$')
ax.grid(True)


plt.show()

print("Mean Final Error: {:0.4f}".format(err_prog[(len(x)-1)]))
print("Variance Final Error: {:0.5f}".format(var_err_prog[(len(x)-1)]))
```

I risultati ottenuti sono compatibili con i valori veri di media e varianza. Per la media si nota come dopo circa 60 simulazioni si ottenga un valore che si allontana di un sigma dal valore atteso; per la varianza ciò avviene dopo circa 50 simulazioni. Gli errori finali sono di 0.0003 per la stima della media e di 0.00007 per la stima della varianza.

Procedo con un controllo del valore di $\chi^2$. Questa quantità è stata valutata per 100 volte consecutive, generando $n=10000$ numeri random che sono stati inseriti in $M=100$ intervalli tra 0 e 1. Nell'ipotesi in cui i valori seguano una distribuzione uniforme, mi aspetto che la probabilità di avere un certo numero $k$ di generazioni in un intervallino segua una distribuzione binomiale, con $\langle k\rangle=np=n/M$ e $\sigma^2 = np(1-p)\simeq np=n/M$. Il $\chi^2$ risulterà quindi
$$
\chi^2 = \sum_{i=1}^M \frac{\left( n_i - n/M \right)^2}{n/M}\simeq100.
$$

Importo i risultati ottenuti e plotto i 100 valori di $\chi^2$ calcolati traslandoli di un valore di 100.

```python
chi_sq = np.loadtxt('output/chi_sq.txt')
chi_sq = chi_sq - 100
```

```python
x = range(len(chi_sq))
plt.figure(figsize=(8,5))
plt.title(r"Evaluation of $\chi^2$", fontsize=14, fontweight="bold")
plt.plot(x,chi_sq)
plt.xlabel('# of Simulations')
plt.ylabel('$\chi^2-100$')
plt.grid(True)
plt.show()
print(r"Mean of chi^2: {:0.3f}".format(100+np.mean(chi_sq)))
```

Dal grafico risulta chiaro come il valore di $\chi^2$ sia compatibile con ciò che mi aspetto da una distribuzione uniforme di dati. Posso concludere che il generatore di numeri casuali sia effettivamente un buon generatore di numeri uniformemente distribuiti.

## Esercizio 01.2
Aggiungo le funzioni per la generazione di numeri distribuiti esponenzialmente
$$p(x) = \lambda \exp(-\lambda x),$$
con $x\in [0;+\infty]$, e secondo una distribuzione di Cauchy-Lorentz
$$p(x)=\frac{1}{\pi}\frac{\Gamma}{(x-\mu)^2+\Gamma^2},$$ $x\in [-\infty;+\infty]$, al generatore di numeri casuali fornito tramite il metodo dell'inversa della cumulata. Ne testo, quindi, la validità.

Ho generato $n=10000$ realizzazioni di 
$$S_N=\frac1N\sum_{i=1}^Nx_i$$
con $N=1,2,10,100$, dove $x_i$ è una variabile stocastica generata uniformemente, esponenzialmente ($\lambda=1$) e secondo una distribuzione di Cauchy-Lorentz ($\mu=0,\Gamma=1$)

Importo i dati generati e ne plotto gli istogrammi:

```python
u = pd.read_csv('output/uniform.txt', sep="\s+", header=None)
u.columns = ["s1", "s2", "s10", "s100"]
exp = pd.read_csv('output/exponential.txt', sep="\s+", header=None)
exp.columns = ["s1", "s2", "s10", "s100"]
CL = pd.read_csv('output/lorentzian.txt', sep="\s+", header=None)
CL.columns = ["s1", "s2", "s10", "s100"]
```

```python
fig, axes = plt.subplots(4,3,figsize = (15,22))
fig.suptitle('Value of $\mathbf{s_N}$', fontsize=16, fontweight="bold")



axes[0,0].hist(x=u.s1, bins=50)
axes[0,1].hist(x=exp.s1, bins=50, density=True)
axes[0,2].hist(x=np.clip(CL.s1, -40, 40), bins=50, density=True)
axes[0,2].set_xlim(-35,35)

mu_exp, lambda_exp = expon.fit(exp.s1)
x = np.linspace(0, 8, 100)
p = expon.pdf(x, mu_exp, lambda_exp)
axes[0,1].plot(x, p, linewidth=1, label = "mu = %.2f,  lambda = %.2f" % (mu_exp, lambda_exp), c="r")
axes[0,1].legend()


mu_CL, std_CL = cauchy.fit(CL.s1)
x = np.linspace(-20, 20, 100)
p = cauchy.pdf(x, mu_CL, std_CL)
axes[0,2].plot(x, p, linewidth=1, label = "mu = %.2f,  std = %.2f" % (mu_CL, std_CL), c="r")
axes[0,2].legend()

axes[1,0].hist(x=u.s2, bins=50)
axes[1,1].hist(x=exp.s2, bins=50)
axes[1,2].hist(x=np.clip(CL.s2, -40, 40), bins=50)
axes[1,2].set_xlim(-35,35)

axes[2,0].hist(x=u.s10, bins=50)
axes[2,1].hist(x=exp.s10, bins=50)
axes[2,2].hist(x=np.clip(CL.s10, -40, 40), bins=50)
axes[2,2].set_xlim(-35,35)


axes[3,0].hist(x=u.s100, bins=50, density=True)
axes[3,1].hist(x=exp.s100, bins=50, density=True)
axes[3,2].hist(x=np.clip(CL.s100, -40, 40), bins=50, density=True)
axes[3,2].set_xlim(-35,35)


axes[0,0].set_ylabel('counts')
axes[0,0].set_title('Uniform Distribution', fontsize=14)
axes[0,1].set_title('Exponential Distribution', fontsize=14)
axes[0,2].set_title('Cauchy-Lorentz Distribution', fontsize=14)  
axes[0,0].set_xlabel('$s_1$')
axes[0,1].set_xlabel('$s_1$')
axes[0,2].set_xlabel('$s_1$')


axes[1,0].set_ylabel('counts')
axes[1,0].set_title('Uniform Distribution', fontsize=14)
axes[1,1].set_title('Exponential Distribution', fontsize=14)
axes[1,2].set_title('Cauchy-Lorentz Distribution', fontsize=14)
axes[1,0].set_xlabel('$s_2$')
axes[1,1].set_xlabel('$s_2$')
axes[1,2].set_xlabel('$s_2$')


axes[2,0].set_ylabel('counts')
axes[2,0].set_title('Uniform Distribution', fontsize=14)
axes[2,1].set_title('Exponential Distribution', fontsize=14)
axes[2,2].set_title('Cauchy-Lorentz Distribution', fontsize=14)
axes[2,0].set_xlabel('$s_{10}$')
axes[2,1].set_xlabel('$s_{10}$')
axes[2,2].set_xlabel('$s_{10}$')

# Fit a normal distribution to the data:
mu_u, std_u = norm.fit(u.s100)
mu_exp, std_exp = norm.fit(exp.s100)
mu_CL, gamma_CL = cauchy.fit(CL.s100)


axes[3,0].set_ylabel('counts')
axes[3,0].set_title('Uniform Distribution', fontsize=14)
axes[3,1].set_title('Exponential Distribution', fontsize=14)
axes[3,2].set_title('Cauchy-Lorentz Distribution', fontsize=14) 
axes[3,0].set_xlabel('$s_{100}$')

#gaussian fit
x = np.linspace(0.4, 0.6, 100)
p = norm.pdf(x, mu_u, std_u)
axes[3,0].plot(x, p, linewidth=1, label = "mu = %.2f,  std = %.2f" % (mu_u, std_u), c="r")
axes[3,0].legend()
axes[3,1].set_xlabel('$s_{100}$')
#gaussian fit
x = np.linspace(0.7, 1.4, 100)
p = norm.pdf(x, mu_exp, std_exp)
axes[3,1].plot(x, p, linewidth=1, label = "mu = %.2f,  std = %.2f" % (mu_exp, std_exp), c="r")
axes[3,1].legend()

axes[3,2].set_xlabel('$s_{100}$')
#cauchy fit
x = np.linspace(-20, 20, 100)
p = cauchy.pdf(x, mu_CL, gamma_CL)
axes[3,2].plot(x, p, linewidth=1, label = "mu = %.2f,  gamma = %.2f" % (mu_CL, gamma_CL), c="r")
axes[3,2].legend()




plt.show()
```

Nella prima serie di istogrammi è verificata la validità dei generatori random, come mostrano i fit eseguiti. Nei successivi istogrammi invece è resa evidente la validità del teorema del limite centrale, che vede realizzazione nell'ultima serie di istogrammi. Per la distribuzione di Cauchy-Lorentz si vede come la distribuzione limite sia una distribuzione di Cauchy-Lorentz, e non una gaussiana come nel caso della distribuzione uniforme e della distribuzione esponenziale.

# Esercizio 01.3
Per valutare il valore di $\pi$ si è fatto ricorso all'esperimento di Buffon: ho "lanciato" $N_{thr}=50000$ aghi (di lunghezza $L=2.5$) su un foglio a righe (di distanza pari a $d=3.0$) ed ho valutato la stima per $\pi$ utilizzando la relazione:
$$\pi = \frac{2L}{Pd} = \lim_{N_{thr} \to \infty}\frac{2LN_{thr}}{N_{hit}d}.$$
Utilizzando la media a blocchi ($M=10000$ valutazioni di $\pi$, divise in $N=100$ blocchi) ne ho estratto una stima statistica.

```python
x, sum_prog, err_prog = np.loadtxt('output/pi.txt', unpack=True)
sum_prog = sum_prog - np.pi
```

```python
plt.figure(figsize=(8,5))

plt.errorbar(x,sum_prog,yerr=err_prog)
plt.title(r"Evaluation of $\pi$", fontsize=14, fontweight="bold")
plt.xlabel('# of Block')
plt.ylabel('$x-\pi$')

plt.grid(True)


plt.show()
print("Final Error: {:0.5f}".format(err_prog[(len(x)-1)]))
```

Dal grafico si vede come già dopo circa 30 simulazioni si sia ottenuto un valore di $\pi$ che si allontana dal valore vero di meno di una sigma. L'errore finale ottenuto è di 0.00012, che ci dà un'ottima stima di $\pi$.
