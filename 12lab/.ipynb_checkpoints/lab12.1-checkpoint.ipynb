```python
import tensorflow as tf
from tensorflow import keras
import os
import numpy as np
import matplotlib.pyplot as plt
seed=0
np.random.seed(seed) # fix random seed
tf.random.set_seed(seed)
```

```python
from tensorflow.keras.datasets import mnist

# input image dimensions
img_rows, img_cols = 28, 28 # number of pixels 
# output
num_classes = 10 # 10 digits

# the data, split between train and test sets
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()


from tensorflow.keras.layers import Flatten, Conv2D, MaxPooling2D

# reshape data, depending on Keras backend
if keras.backend.image_data_format() == 'channels_first':
    X_train = X_train.reshape(X_train.shape[0], 1, img_rows, img_cols)
    X_test = X_test.reshape(X_test.shape[0], 1, img_rows, img_cols)
    input_shape = (1, img_rows, img_cols)
else:
    X_train = X_train.reshape(X_train.shape[0], img_rows, img_cols, 1)
    X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols, 1)
    input_shape = (img_rows, img_cols, 1)

print('X_train shape:', X_train.shape)
print('Y_train shape:', Y_train.shape)
print()
print(X_train.shape[0], 'train samples')
print(X_test.shape[0], 'test samples')
```

```python
# reshape data, it could depend on Keras backend
X_train = X_train.reshape(X_train.shape[0], img_rows*img_cols)
X_test = X_test.reshape(X_test.shape[0], img_rows*img_cols)

# cast floats to single precesion
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

# rescale data in interval [0,1]
X_train /= 255
X_test /= 255

# look at an example of data point
print('an example of a data point with label', Y_train[20])
# matshow: display a matrix in a new figure window
plt.matshow(X_train[20,:].reshape(28,28),cmap='binary')
plt.show()

print('an example of a data point with label', Y_train[20], 'before to_categorical ...')
# convert class vectors to binary class matrices, e.g. for use with categorical_crossentropy
Y_train = keras.utils.to_categorical(Y_train, num_classes)
Y_test = keras.utils.to_categorical(Y_test, num_classes)
print('... and with label', Y_train[20], 'after to_categorical')
print()
print('X_train shape:', X_train.shape)
print('Y_train shape:', Y_train.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

def create_DNN():
    # instantiate model
    model = Sequential()
    # add a dense all-to-all relu layer
    model.add(Dense(400,input_shape=(img_rows*img_cols,), activation='relu'))
    # add a dense all-to-all relu layer
    model.add(Dense(100, activation='relu'))
    # apply dropout with rate 0.5
    model.add(Dropout(0.5))
    # soft-max layer
    model.add(Dense(num_classes, activation='softmax'))
    
    return model

print('Model architecture created successfully!')
```

```python
from tensorflow.keras.optimizers import SGD, Adam, RMSprop, Adagrad, Adadelta, Adamax, Nadam

def compile_model(optim):
    # create the model
    model=create_DNN()
    # compile the model
    model.compile(loss=keras.losses.categorical_crossentropy,
                  optimizer=optim,
                  metrics=['accuracy'])
    return model

print('Model compiled successfully and ready to be trained.')
```

```python
optimizer = ["sgd","adam","rmsprop","adagrad","adadelta","adamax","nadam"]

history = []

Train_accuracy = []      #Define matrices to store accuracy scores as a function
Test_accuracy = []       #of n_epochs and number of hidden neurons for 

# training parameters
batch_size = 32
epochs = 5

for i in range(len(optimizer)):     #run loops over hidden neurons and learning rates to calculate 
    DNN_model = compile_model(optimizer[i])
    history.append(DNN_model.fit(X_train, Y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(X_test, Y_test)))
    
    Train_accuracy.append(DNN_model.evaluate(X_train,Y_train)[1])
    Test_accuracy.append(DNN_model.evaluate(X_test,Y_test)[1])
    
    
```

```python
def plot_data_optim(data,title):

    # plot results
    fontsize=16
    
    fig = plt.figure(figsize = (14,7))
    ax = fig.add_subplot(111)

    ax.set_title(title + " Accuracy", fontsize=fontsize,fontweight='bold')
    
    colors = ["orangered","gold","olive","limegreen","cadetblue",#"steelblue",
              "mediumpurple","mediumorchid"]
    
    x = np.arange(len(optimizer))
    y = data
    ax.set_xticklabels(['']+optimizer)
    ax.bar(x, y, color=colors)
    for i, v in enumerate(y):
        ax.text( i-0.17,v + 0.03, str("${0:.2f}\%$".format(v*100)),va = "center")    
    
    
    ax.set_ylabel('Loss', fontsize = fontsize - 2, fontweight = 'bold')
    ax.set_xlabel('Optimizer', fontsize = fontsize - 2, fontweight = 'bold')
    
    plt.show()

plot_data_optim(Train_accuracy, 'Training')
plot_data_optim(Test_accuracy, 'Testing')
```

Mi concentro sui quattro ottimizzatori che hanno dato i risultati migliori (*adam*, *rmsprop*, *adamax* e *nadam*) ed aumento a $n_{epochs}=10$ il numero di epoche svolte.

```python
optimizer_best = ["adam","rmsprop","adamax","nadam"]

history_best = []
score_best = []
predictions = []

# training parameters
batch_size = 32
epochs = 10

for i in range(len(optimizer_best)):     #run loops over hidden neurons and learning rates to calculate 
    DNN_model = compile_model(optimizer_best[i])
    history_best.append(DNN_model.fit(X_train, Y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(X_test, Y_test)))
    score_best.append(DNN_model.evaluate(X_test, Y_test, verbose=1))
    predictions.append(DNN_model.predict(X_test))

```

```python
import random
from random import sample
def plot_loss_accuracy_opt_predictions(optimizer,history,score,testx,testy,predict):
    testx = testx.reshape(testx.shape[0], img_rows, img_cols,1)
    rnd = random.sample(range(len(testx)),10)
    x = range(0,10)
    
    # print performance
    for i in range(len(history)):
        
        fig, ax = plt.subplots(1, 2, figsize=(17, 5))
        fig.suptitle("NN Model\n" + optimizer[i], fontsize=16, fontweight = "bold")
        
        ax[0].plot(x,history[i].history['accuracy'])
        ax[0].plot(x,history[i].history['val_accuracy'])
        ax[0].set_title('Model accuracy')
        ax[0].set_ylabel('Accuracy')
        ax[0].set_xlabel('Epoch')
        ax[0].legend(['Train', 'Test'], loc='best')
        
        ax[1].plot(x,history[i].history['loss'])
        ax[1].plot(x,history[i].history['val_loss'])
        ax[1].set_title('Model loss')
        ax[1].set_ylabel('Loss')
        ax[1].set_xlabel('Epoch')
        ax[1].legend(['Train', 'Test'], loc='best')
                
        plt.show()

        print()
        print('Test accuracy:', score[i][1])
        print('Test loss:', score[i][0])
        
        plt.figure(figsize=(15, 5)) 
        for j in range(10):    
            ax = plt.subplot(2, 10, j + 1)    
            plt.imshow(testx[rnd[j], :, :, 0], cmap='gray')    
            plt.title("Digit: {}\nPredicted: {}".format(np.argmax(testy[rnd[j]]), np.argmax(predict[i][rnd[j]])))    
            plt.axis('off') 
        plt.show()
        
plot_loss_accuracy_opt_predictions(optimizer_best, history_best, score_best, X_test, Y_test, predictions)
```

Guardando all'accuracy delle reti neurali, si pu notare come sia con l'ottimizzatore *Adam*, sia con l'ottimizzatore *NAdam* ci sia una tendenza all'overfitting: in entrambi i casi vediamo un progressivo miglioramento delle prestazioni sul set di training a cui non corrisponde un miglioramento alle prestazioni sul set di test. Questa tendenza sembra rivelarsi anche nel caso dell'ottimizzatore *Adamax*. L'ottimizzatore *RMSprop*, invece, sembra evitare l'overfitting e per questo motivo sembra essere la soluzione pi azzeccata per un'analisi pi approfondita.
