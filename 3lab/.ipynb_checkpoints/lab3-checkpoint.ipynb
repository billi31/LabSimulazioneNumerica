## Esercizio 03.1
In questo esercizio valuto con una simulazione Monte Carlo i prezzi di CALL-option e PUT-option europee al tempo $t=0$, considerando che
$$ C[S(0),0]=e^{-rT}\max\left[0,S(T)-K\right]$$
e
$$ P[S(0),0]=e^{-rT}\max\left[0,K-S(T)\right],$$
dove $S(T)$ è l'asset price al tempo finale $T$, $r$ è il tasso di interesse e $K$ è lo strike price (prezzo al quale il possessore dell'opzione può esercitare la facoltà). Per il calcolo dell'asset price al tempo $T$, utilizzo due diversi metodi: 
- calcolo direttamente $S(T)$ a partire da $S(0)$ 
$$S(T) = S(0) \exp\left[ \left(r-\frac{1}{2}\sigma^2\right)t+\sigma W(t)\right],$$
dove $W(t)\sim\mathcal{N}(0,t)$ e $\sigma$ è la volatilità;
- calcolo $S(T)$ come risultato di un moto Browniano geometrico, sapendo che
$$S(t_{i+1}) = S(t_i) \exp\left[ \left(r-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)+\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right],$$
dove $Z_i \sim \mathcal{N}(0,1)$ e $\sigma$ è la volatilità.

Confronto le soluzioni numeriche trovate con i valori analitici della teoria di pricing di Black-Scholes.

Importo i risultati ottenuti dalla simulazione del prezzo di CALL-option e di PUT-option, calcolati tramite valutazione diretta e a step dell'asset price $S(T)$, con i parametri: 
- asset price at $t=0$: $S(0)=100$
- delivery time: $T=1$
- strike price: $K=100$
- risk-free interest rate: $r=0.1$
- volatility: $\sigma=0.25$

Nella valutazione a step ho diviso l'intervallo di tempo $[0,1]$ in $n_{passi}=100$; per la statistica ho utilizzato la media a blocchi, suddividendo le $M=100000$ valutazioni di $S(T)$ in $N=100$ blocchi.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.optimize as opt
sns.set()
```

```python
x_C, sum_prog_C, err_prog_C = np.loadtxt('output/CALL.txt', unpack=True)
x_CD, sum_prog_CD, err_prog_CD = np.loadtxt('output/CALL_d.txt', unpack=True)

x_P, sum_prog_P, err_prog_P = np.loadtxt('output/PUT.txt', unpack=True)
x_PD, sum_prog_PD, err_prog_PD = np.loadtxt('output/PUT_d.txt', unpack=True)
```

```python
from scipy.special import erf

# data
S0 = 100.
K = 100.
T = 1.
r = 0.1
sigma = 0.25
t=0

def N(x):
    return 0.5 * (1. + erf(x / np.sqrt(2.)))

def black_scholes(S0, K, T, r, sigma):
    d1 = 1./(sigma * np.sqrt(T)) * (np.log(S0 / K) + (r + (sigma**2) / 2.) * T)
    d2 = d1 - sigma * np.sqrt(T)
    C = S0 * N(d1) - K * np.exp(-r * T) * N(d2)
    P = S0 *(N(d1) - 1.) - K * np.exp(-r * T) * (N(d2)-1.)
    return C, P

if __name__ == "__main__":
    call, put = black_scholes(S0, K, T, r, sigma)
    print(" == BLACK-SCHOLES ==")
    print("call: ", call)
    print("put: ", put)
```

Confronto i risultati con la soluzione analitica di Black-Scholes:

```python
fig = plt.figure(figsize = (17,5))

fig.suptitle("CALL-option Price", fontsize = 16, fontweight = "bold")
ax = fig.add_subplot(121)
ax.errorbar(x_C,sum_prog_C,yerr=err_prog_C)
ax.set_title(r'Direct Calculation',fontsize = 14, fontweight = "bold")

horiz_line_data = np.array([call for i in range(len(sum_prog_C)+1)])
ax.plot(horiz_line_data, 'r--')

ax.set_xlabel(r'# of Block', fontweight = "bold")
ax.set_ylabel(r'CALL-option Price', fontweight = "bold")
ax.grid(True)

ax = fig.add_subplot(122)
ax.errorbar(x_CD,sum_prog_CD,yerr=err_prog_CD)
ax.set_title(r'Discretized Calculation',fontsize = 14, fontweight = "bold")
ax.plot(horiz_line_data, 'r--')


ax.set_xlabel(r'# of Block', fontweight = "bold")
ax.set_ylabel(r'CALL-option Price', fontweight = "bold")
ax.grid(True)

plt.show()

fig = plt.figure(figsize = (17,5))

fig.suptitle("PUT-option Price", fontsize = 16, fontweight = "bold")
ax = fig.add_subplot(121)
ax.errorbar(x_P,sum_prog_P,yerr=err_prog_P)
ax.set_title(r'Direct Calculation',fontsize = 14, fontweight = "bold")

horiz_line_data = np.array([put for i in range(len(sum_prog_P)+1)])
ax.plot(horiz_line_data, 'r--')

ax.set_xlabel(r'# of Block', fontweight = "bold")
ax.set_ylabel(r'PUT-option Price', fontweight = "bold")
ax.grid(True)

ax = fig.add_subplot(122)
ax.errorbar(x_PD,sum_prog_PD,yerr=err_prog_PD)
ax.set_title(r'Discretized Calculation',fontsize = 14, fontweight = "bold")
ax.plot(horiz_line_data, 'r--')


ax.set_xlabel(r'# of Block', fontweight = "bold")
ax.set_ylabel(r'PUT-option Price', fontweight = "bold")
ax.grid(True)

plt.show()
```

Tutti e quattro i grafici mostrano compatibilità con il valore vero (secondo Black-Scholes) di pricing, infatti i risultati ottenuti sono lontani meno di una $\sigma$ dal valore atteso.

```python
fig, axes = plt.subplots(figsize = (9,5))
plt.plot(x_C,err_prog_C, label = "CALL: Direct Evaluation")
plt.plot(x_CD,err_prog_CD,'r', label = "CALL: Step Evaluation")

plt.plot(x_P,err_prog_P, label = "PUT: Direct Evaluation")
plt.plot(x_PD,err_prog_PD, label = "PUT: Step Evaluation")


plt.title(r'Price Extimation Error',fontsize = 14, fontweight = "bold")

plt.xlabel('# of Block', fontweight = "bold")
plt.ylabel('Error', fontweight = "bold")
plt.legend()
plt.show()


print("CALL-option Direct Final Error: {:0.5f}".format(err_prog_C[(len(x_C)-1)]))
print("CALL-option Step   Final Error: {:0.6f}".format(err_prog_CD[(len(x_CD)-1)]))
print("PUT-option  Direct Final Error: {:0.5f}".format(err_prog_P[(len(x_P)-1)]))
print("PUT-option  Step   Final Error: {:0.6f}".format(err_prog_PD[(len(x_PD)-1)]))
```

Confrontando gli errori non si nota grande differenza tra la valutazione diretta o a step: in generale, però, sembra meglio utilizzare la valutazione diretta dell'asset price.
